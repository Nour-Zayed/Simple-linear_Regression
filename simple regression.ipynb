```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score ,mean_squared_error
import os
```

```python
os.chdir(r"C:\Users\NooR\Downloads\machine learning\simple liner regression")
data= pd.read_csv("salary_data.csv")
## first / last 5 lines  in data
data.head()
data.tail()
#some info
data.info()
#some statistic use with numerical data 
data.describe()
sns.pairplot(data)
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,1]
```

```python


x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)

my_model=LinearRegression()
my_model.fit(x_train,y_train)
my_model.score(x_train,y_train)

```

```python
plt.scatter (x_train,y_train ,color='red')
plt.plot (x_train , my_model.predict(x_train) ,color='blue')
plt.title ("salary vs experience (train set)")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```

```python
y_pred=my_model.predict(x_test)
plt.scatter(x_test,y_test ,color='red')
plt.plot(x_test ,y_pred, color='blue')
plt.title ("salary vs experience (test set)")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```

```python
# Plotting the actual and predicted values
c=[i for i in range (1,len(y_test)+1)]
plt.plot (c,y_test,color='red',linestyle='-')
plt.plot (c,y_pred ,color='blue' ,linestyle='-')
plt.title ("salary vs experience (test set)")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```

```python
## plotting the error
c=[i for i in range (1,len(y_test)+1)]
plt.plot(c,y_test-y_pred,color='green',linestyle='-')
plt.xlabel('index')
plt.ylabel('Error')
plt.title('Error Value')
plt.show()
```

```python
# calculate Mean square error
mse=mean_squared_error(y_test,y_pred)

# Calculate R square vale
rsq=r2_score(y_test,y_pred)
print("mean squre error =" , mse)
print("r squre =" , rsq)
```
